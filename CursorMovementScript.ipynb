```python
import time
import random
import pyautogui as pag
import keyboard

pag.FAILSAFE = True

try: 
    while True:
        if keyboard.is_pressed("q"):
            print("Stop the script")
            break
        x = 50
        y = 1050
        pag.moveTo(x,y,duration=0.5)
        pag.click()
        time.sleep(30)
except pag.FailSafeException:
    print("Cursor moved to top left corner. Stop the script")
except KeyboardInterrupt:
    print("Stop script using keyboard")
        
        
```
